```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_iris
data = load_iris()
data
```

```python
sepal_length = []
sepal_width = []
petal_length =[]
petal_width =[]
class1 = []

for i in range(len(data['data'])):
  sepal_length.append(data['data'][i][0])
  sepal_width.append(data['data'][i][1])
  petal_length.append(data['data'][i][2])
  petal_width.append(data['data'][i][3])
  class1.append(data['target_names'][data['target'][i]])

data_dict = {'sepal_length':sepal_length, 'sepal_width':sepal_width, 'petal_length':petal_length, 'petal_width':petal_width, 'class':class1}
df = pd.DataFrame(data_dict)
df.head()
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
X=df.drop('class', axis=1)
y=df['class']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.20)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)

print("Score:",dt.score(X_test, y_test))
```

```python
from sklearn.metrics import accuracy_score

y_pred = dt.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)*100

print("Accuracy:",accuracy)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt

fig, axis = plt.subplots(figsize=(30,20))
tree.plot_tree(dt, feature_names=X.columns, class_names=y.unique(), filled=True, ax=axis)
plt.show()
```
