```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/HousingData.csv')
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['CRIM'].fillna(df['CRIM'].mean(), inplace=True)
df['CRIM'].mean()
```

```python
df['CRIM'].isnull().sum()
```

```python
df['ZN'].fillna(df['ZN'].mean(), inplace=True)
df['INDUS'].fillna(df['INDUS'].mean(), inplace=True)
df['AGE'].fillna(df['AGE'].mean(), inplace=True)
df['LSTAT'].fillna(df['LSTAT'].mean(), inplace=True)
```

```python
df.CHAS.value_counts()
```

```python
df.CHAS.fillna(method='bfill', inplace=True) # backfill / bfill: use next valid observation to fill gap.
df.CHAS.value_counts()
```

```python
df.isnull().sum()
```

```python
X=df.iloc[:,0:-1]
X.head()
```

```python
y=df.iloc[:,-1]
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=42, test_size=0.20)
```

```python
from sklearn.tree import DecisionTreeRegressor #Use DT Regressor for continuous target/response variable

dt = DecisionTreeRegressor(max_depth=6)
dt.fit(X_train, y_train)

dt.score(X_test, y_test)
```

```python
y_pred = dt.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_pred, y_test) # accuracy_score can't be used for continous response variables
print("Accuracy:", accuracy)
```

```python
print(y_pred.mean())
print(y_test.mean())
```

```python
import matplotlib.pyplot as plt
from sklearn import tree

fig, axis = plt.subplots(figsize=(40,10)) # figsize(width, len)?
tree.plot_tree(dt, feature_names=X_train.columns, class_names=y_train.unique(), filled=True, ax=axis, fontsize=10)
plt.show()
```
